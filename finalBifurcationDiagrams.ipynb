```python
import pandas as pd
import matplotlib.pyplot as plt
file_path = r"C:\xppall\final.dat"
plt.rcParams.update({'font.size': 14,'lines.linewidth': 3}) 

df = pd.read_csv(file_path, delimiter="\s+", header=None)
df = df.iloc[1:,:]
Istim = df.iloc[:, 0]  # First column: control parameter
v = df.iloc[:, 1]  # Second column: bifurcating variable
stability_col = df.iloc[:, 3] 
first_stable =  df[(stability_col == 1 )& (Istim<500)]
unstable = df[(stability_col == 2 )]#
second_stable =  df[(stability_col == 1 )& (Istim>500)]
remaining_data = df[~df.index.isin(first_stable.index) & 
                    ~df.index.isin(unstable.index) & 
                    ~df.index.isin(second_stable.index)]
lim_cyc_up = remaining_data.iloc[:,1]
lim_cyc_down = remaining_data.iloc[:,2]
plt.plot(first_stable.iloc[:,0],first_stable.iloc[:,1],'k',label = "Stable Equilibrium")
plt.plot(second_stable.iloc[:,0],second_stable.iloc[:,1],'k')
plt.plot(unstable.iloc[:,0],unstable.iloc[:,1],'r',label = "Unstable Equilibrium")
plt.plot(remaining_data.iloc[:,0],lim_cyc_up,'g--')
plt.plot(remaining_data.iloc[:,0],lim_cyc_down,'g--',label = "Stable Limit Cycle")
plt.xlim([-50, 1000])
plt.ylim([-100, 40])
plt.legend()

plt.xlabel(r"$I_{stim} (\mu A/cm^2)$")
plt.ylabel(r"$V (mV)$")
plt.show()

plt.plot(first_stable.iloc[:,0],first_stable.iloc[:,1],'k',label = "Stable Equilibrium")
plt.plot(second_stable.iloc[:,0],second_stable.iloc[:,1],'k')
plt.plot(unstable.iloc[:,0],unstable.iloc[:,1],'r',label = "Unstable Equilibrium")
plt.plot(remaining_data.iloc[:,0],lim_cyc_up,'g--')
plt.plot(remaining_data.iloc[:,0],lim_cyc_down,'g--',label = "Stable Limit Cycle")
plt.xlim([0, 80])
plt.ylim([-80, 40])
plt.legend()
plt.xlabel(r"$I_{stim} (\mu A/cm^2)$")
plt.ylabel(r"$V (mV)$")
plt.show()
```
